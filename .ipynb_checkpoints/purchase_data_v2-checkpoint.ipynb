```python
import pandas as pd

df = pd.read_csv('purchase_data.csv')
df
```

```python
totalplayers=df["SN"].nunique()
totalplayers
```

```python

numberofuniqueitems = df["Item ID"].nunique()
avgpurchaseprice = df["Price"].mean()
totalnumberofpurchases = df["Purchase ID"].count()
totalrevenue = df["Price"].sum()

                              
pa = ({"number of unique items":[numberofuniqueitems],
       "average purchase price":[avgpurchaseprice],
       "total number of purchases":[totalnumberofpurchases],
       "total revenue":[totalrevenue]})
    

Analysis_df = pd.DataFrame([numberofuniqueitems,avgpurchaseprice,totalnumberofpurchases,totalrevenue])
PurchaseAnalysis_df = Analysis_df.T
PurchaseAnalysis_df.columns = ["number of unique items","average purchase price","total number of purchases","total revenue"]
PurchaseAnalysis_df
```

```python
# maleplayers = df[df["Gender"] == "Male"]
# maleplayercount = maleplayers["SN"].nunique()
# femaleplayers = df[df["Gender"] == "Female"]
# femaleplayercount = femaleplayers["SN"].nunique()                    
# otherplayers = df[df["Gender"] == "Other / Nondisclosed"]
# otherplayercount = otherplayers["SN"].nunique()
                     
# gd = pd.DataFrame
                     
# "maleplayercount":[maleplayercount],
# "femaleplayercount":[femaleplayercount],
# "otherplayercount":[otherplayercount],

gendertotal = df.groupby("Gender")["SN"].nunique()
pct = (gendertotal / totalplayers) * 100

gd = pd.DataFrame({"Total Count": gendertotal, "Percent of Players": pct})
gd["Percent of Players"] = gd["Percent of Players"].map("{:.2f}%".format)
gd
```

```python
output=df.groupby('Gender').agg({'Price':['count','mean','sum']})
## This calculates the avg total person
output.loc[:,('Price','Avg total per person')]=output.loc[:,('Price','sum')]/df.groupby("Gender")["SN"].nunique()
output
```

```python
## - {} --> dictionary {key:value , key:value}
## - [] --> List
## - () --> tupple
```

```python
## Rename the column names
output=output.rename(columns={'Price':'Output','count':'Purchase Count','mean':'Average Purchase Price','sum':'Total Purchase value'})
```

```python
output
```

```python
## Transform to two decimals and add dollar symbol ($)
import numpy as np
output.loc[:,('Output','Average Purchase Price')]=output.loc[:,('Output','Average Purchase Price')].apply(lambda x: '$'+str(np.round(x,2)))
output.loc[:,('Output','Total Purchase value')]=output.loc[:,('Output','Total Purchase value')].apply(lambda x: '$'+str(np.round(x,2)))
output.loc[:,('Output','Avg total per person')]=output.loc[:,('Output','Avg total per person')].apply(lambda x: '$'+str(np.round(x,2)))
```

```python
## apply is a method used on Series, we apply one function to all the entries of the Series
```

## Age Demographics

```python
#age_demographics = df.groupby(bins)[['SN']].agg('nunique').rename(columns={'SN':'Total Count'})
```

```python
bins=pd.cut(df['Age'],[0,9,14,19,24,29,34,39,100])
output1=df.groupby(bins)[['SN']].agg('nunique').rename(columns={'SN':'Total Count'})
output1['Percentage of Players']=(output1['Total Count']/576*100).apply(lambda x: str(np.round(x,2))+'%')
output1.index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
#output1['Percentage of Players']=output1['Percentage of Players']
```

## Purchasing Analysis (Age)

```python
output2=df.groupby(bins).agg({'Price':['count','mean','sum']})
output2
```

```python
## Calculate Average total per person on each range: Total Purchase value / number of unique people per range
output2.loc[:,('Price','Avg total per person')]=output2.loc[:,('Price','sum')]/df.groupby(bins)["SN"].nunique()
```

```python
## Rename column names
output2.rename(columns={'Price':'Output','count':'Purchase Count','mean':'Average Purchase Price','sum':'Total Purchase value'},inplace=True)

## Change values to two decimals and add dollar symbol ($)
output2.loc[:,('Output','Average Purchase Price')]=output2.loc[:,('Output','Average Purchase Price')].apply(lambda x: '$'+str(np.round(x,2)))
output2.loc[:,('Output','Total Purchase value')]=output2.loc[:,('Output','Total Purchase value')].apply(lambda x: '$'+str(np.round(x,2)))
output2.loc[:,('Output','Avg total per person')]=output2.loc[:,('Output','Avg total per person')].apply(lambda x: '$'+str(np.round(x,2)))
```

```python
## Rename the index values
output2.index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
```
