```python
output=df.groupby(by='Gender').agg({'Price':['count','mean','sum']})
## This calculates the avg total person
output.loc[:,('Price','Avg total per person')]=output.loc[:,('Price','sum')]/df.groupby("Gender")["SN"].nunique()
```

```python
## - {} --> dictionary {key:value , key:value}
## - [] --> List
## - () --> tupple
```

```python
## Rename the column names
output=output.rename(columns={'Price':'Output','count':'Purchase Count','mean':'Average Purchase Price','sum':'Total Purchase value'})
```

```python
output
```

```python
## Transform to two decimals and add dollar symbol ($)
import numpy as np
output.loc[:,('Output','Average Purchase Price')]=output.loc[:,('Output','Average Purchase Price')].apply(lambda x: '$'+str(np.round(x,2)))
output.loc[:,('Output','Total Purchase value')]=output.loc[:,('Output','Total Purchase value')].apply(lambda x: '$'+str(np.round(x,2)))
output.loc[:,('Output','Avg total per person')]=output.loc[:,('Output','Avg total per person')].apply(lambda x: '$'+str(np.round(x,2)))
```

```python
## apply is a method used on Series, we apply one function to all the entries of the Series
```

## Purchasing Analysis (Age)

```python
output2=df.groupby(bins).agg({'Price':['count','mean','sum']})
output2
```

```python
## Calculate Average total per person on each range: Total Purchase value / number of unique people per range
output2.loc[:,('Price','Avg total per person')]=output2.loc[:,('Price','sum')]/df.groupby(bins)["SN"].nunique()
```

```python
## Rename column names
output2.rename(columns={'Price':'Output','count':'Purchase Count','mean':'Average Purchase Price','sum':'Total Purchase value'},inplace=True)

## Change values to two decimals and add dollar symbol ($)
output2.loc[:,('Output','Average Purchase Price')]=output2.loc[:,('Output','Average Purchase Price')].apply(lambda x: '$'+str(np.round(x,2)))
output2.loc[:,('Output','Total Purchase value')]=output2.loc[:,('Output','Total Purchase value')].apply(lambda x: '$'+str(np.round(x,2)))
output2.loc[:,('Output','Avg total per person')]=output2.loc[:,('Output','Avg total per person')].apply(lambda x: '$'+str(np.round(x,2)))
```

```python
## Rename the index values
output2.index=['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
```
